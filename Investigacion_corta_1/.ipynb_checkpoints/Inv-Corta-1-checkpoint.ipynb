# Exploratory Data Analysis (EDA) using Pandas

This notebook uses a simple dataset to explain some basic process in the EDA process, using Pandas and Python

```python
# Import stuff
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from numpy import array
```

The first part is to load the Dataset to be used. In this case, the data consists to the salaries
of the San Francisco Controller's office

```python
# Load the dataset. It corresponds to The San Francisco Controller's Office salaries
# The data format is:
# Year Type,Year,Organization Group Code,Organization Group,Department Code,Department,Union Code,Union,Job Family Code,Job Family,Job Code,Job,Employee Identifier,Salaries,Overtime,Other Salaries,Total Salary,Retirement,Health and Dental,Other Benefits,Total Benefits,Total Compensation
dataset = pd.read_csv("Employee_Compensation.csv", sep=',')
print("INFO: Dataset loaded correctly")
```

Now, it is important to generate some plots, to visually see the data
To do it, we are using MatPlotLib

```python
# Generate some graphics to understand better the dataset

# Get the year column (second column of the dataset)
year = dataset.iloc[:, 1]

# Get the total compensation column (last column of the dataset)
total_compensation = dataset.iloc[:, -1]

# Generate the plot of the total compensation vs the year
fig, ax = plt.subplots()
ax.plot(year, total_compensation, 'ro')

ax.set(xlabel='year', ylabel='total compensation (USD)',
       title='Total compensation vs Year')
ax.grid()
plt.show()

# Generate an histogram of the Year column
plt.hist(dataset.Year)
plt.show()
```

Now, we can see that there is some outlied values... 
So, the first part of the clean-up process is to remove them using some Pandas magic :P
    
The first part is to define the lower and upper bounds of the quantiles.
Then, create a boolean list indicating if the value is in the list or not. 
After that, calculate the median and fill the outlied values with it. At the end you can see in the new graph, that the outlined values were fixed.

```python
# In a year basis, we can see that there is one outlied value (year=2028, that is in the future BTW)
# So, the first clean up to do is to remove the outlied values
print("INFO: Removing outlied values")
lower_bound = 0.01
upper_bound = 0.98

quantile = dataset.Year.quantile([lower_bound, upper_bound])
print("Quantile limits to the year column: \n{}".format(quantile))

print("INFO: Calculating the True index")
true_index = [quantile.loc[lower_bound] < year and year < quantile.loc[upper_bound] for year in dataset.Year.values]
false_index = np.logical_not(true_index)

print("INFO: Calculating the median")
mid = np.median(dataset.Year[true_index])
print("The median is: {}".format(mid))

dataset.Year[false_index] = mid

# Generate a new histogram of the Year column
plt.hist(dataset.Year)
plt.show()
```

The second part of this EDA is to deal with the missing values. What can we do?
Well, we have some options like drop the empty columns, drop the rows that has NAs (empty value), etc. 

Check out the following cell:

```python
print("INFO: Dealing with missing values")

# Drop the columns where all the elements are missing values
print("INFO: Dropping the empty columns")
dataset.dropna(axis=1, how='all')

# Drop the rows where there is more than 10% of missing values
print("INFO: Dropping the columns with more than 10% of missing values")
threshold = int(len(dataset.columns)*0.1)
print("The max number of missing values is: {}".format(threshold))
dataset.dropna(thresh=threshold)

# Finally, fill the remaining values with the mean of the column
print("INFO: Filling the remaining values with the mean")
# IMPORTANT: Since the dataset is too large (~1M of rows and 22 columns), the following operation could take too long to complete
# dataset.fillna(dataset.mean())

print("DONE")
```

Finally, we can generate some plots to analyze the relations between the data. In this dataset we have 22 columns and ~800k rows, so we need to be carefull with the data that we plot, because we can run out of computing resources. 

As an example, I'm generating two graphs: Salaries vs retirement subsidy and the total compensation vs the job,  because we expect them to be proportional.

```python
print("INFO: Plotting different values to analyze relations")

# Salaries vs retirement subsidy plot
x = dataset.Retirement.values
y = dataset.Salaries.values
plt.scatter(x, y, alpha=0.5)
plt.title('Salaries vs retirement subsidy')
plt.xlabel('Retirement')
plt.ylabel('Salaries')
plt.show()

# Total compensation vs job plot
# Get the total compensation column (last column of the dataset)
total_compensation = dataset.iloc[:, -1]
# Get the job column (10th column of the dataset)
job = dataset.iloc[:, 10]
fig, ax = plt.subplots()
ax.plot(job, total_compensation, 'ro')
ax.set(xlabel='job', ylabel='total compensation (USD)',
       title='Total compensation vs Job')
ax.grid()
plt.show()
```
